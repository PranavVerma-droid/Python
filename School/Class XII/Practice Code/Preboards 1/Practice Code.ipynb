```python
cursor.execute("CREATE DATABASE school")
cursor.execute("USE school")

cursor.execute("CREATE TABLE students ( \
Rollno      INTEGER(3)      NOT NULL PRIMARY KEY, \
Name        VARCHAR(20))")

cursor.execute("SHOW TABLES")

print("Tables: ")
for i in cursor:
    print(i)
print()

print("'student' Table Structure: ")
cursor.execute("DESCRIBE students")

for i in cursor:
    print(i)
print()

cursor.execute("DROP TABLE students")
cursor.execute("DROP DATABASE school")
```

```python
cursor.execute("CREATE DATABASE IF NOT EXISTS school")
cursor.execute("USE school")

cursor.execute("CREATE TABLE IF NOT EXISTS students ( \
Rollno      INTEGER(3)      NOT NULL PRIMARY KEY, \
Name        VARCHAR(20))")

cursor.execute("ALTER TABLE students MODIFY Rollno VARCHAR(3)")

print("'student' Table Structure: ")
cursor.execute("DESCRIBE students")

for i in cursor:
    print(i)
print()

mydb.commit()
print(cursor.rowcount)

cursor.execute("DROP TABLE students")
cursor.execute("DROP DATABASE school")

mydb.commit()
```

## Data File Handling

```python
f = open("Files/test.txt", "r+")

data = f.readlines()
print(data)

f.write(". YES IT DOES!")

f.seek(0)
data = f.readlines()
print(data)
f.close()
```

```python
with open("Files/test.txt", "r+") as f:
    data = f.readlines()
    print(data)

    f.write(" YES IT DOES!")

    f.seek(0)
    data = f.readlines()
    print(data)
```

```python
import pickle

f = open("Files/data.dat", "wb")

l1 = [12, 13, 14, 15, 16]

pickle.dump(l1, f)
pickle.dump(l1, f)
pickle.dump(l1, f)
pickle.dump(l1, f)
f.close()

f1 = open("Files/data.dat", "rb")
while True:
    try:
        data = pickle.load(f1)
        print(data)
        print("Tell Value: " + str(f1.tell()))
    except EOFError:
        break

f1.close()
```
